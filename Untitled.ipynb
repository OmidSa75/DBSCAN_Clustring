```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
centers = np.array([[0.5, 2], [-1, -1], [1.5, -1]])

x, y = make_blobs(n_samples=500, centers=centers, cluster_std=0.5, random_state=0)

x = StandardScaler().fit_transform(x)
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='Paired')
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.4, min_samples=20)
db.fit(x)
```

```python
y_pred = db.fit_predict(x)

plt.figure(figsize=(15, 10))
plt.scatter(x[:, 0], x[:, 1], c=y_pred, cmap='Paired')
plt.title('Clusters determined by DBSCAN')
```

```python
from sklearn.datasets import make_gaussian_quantiles
```

```python
x0, y0 = make_gaussian_quantiles(mean=(0, 0), n_samples=500,n_classes=1)
x1, y1 = make_gaussian_quantiles(mean=(4, 4), n_samples=500,n_classes=1)
y1[:] = 1
x2, y2 = make_gaussian_quantiles(mean=(-5, 5), n_samples=500,n_classes=1)
y2[:] = 2

x = np.concatenate((x0, x1, x2))
y = np.concatenate((y0, y1, y2))

x = StandardScaler().fit_transform(x)

plt.figure(figsize=(15, 10))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='Paired')
plt.show()
```

```python
db = DBSCAN(eps=0.2, min_samples=20)
db.fit(x)

y_pred = db.fit_predict(x)

plt.figure(figsize=(15, 10))
plt.scatter(x[:, 0], x[:, 1], c=y_pred, cmap='Paired')
plt.title('Clusters determined by DBSCAN')
```
